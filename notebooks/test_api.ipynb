```python
import requests
from pprint import pprint

BASE_URL = "http://localhost:5000/api"

def test_product_endpoint():
    params = {
        'id': 'c2191318-a71a-4d2d-9708-1c1a184941d3',
        'language': 'es',
        'country': 'ES',
        'currency': 'EUR'
    }
    response = requests.get(f"{BASE_URL}/product", params=params)
    print("Status Code:", response.status_code)
    pprint(response.json())

def test_search_endpoint():
    params = {
        'query': 'camisas',
        'page': 1,
        'limit': 10,
        'sortBy': 'price',
        'ascending': 'true',
        'threshold': 0.8,
        'maxPrice': 100,
        'minPrice': 10,
        'category': 'women',
        # 'onSale': 'true',
        # 'tags': 'linen,white',
        'brands': 'Mango,Zara',
        'language': 'es',
        'country': 'ES',
        'currency': 'EUR'
    }
    response = requests.get(f"{BASE_URL}/search", params=params)
    print("Status Code:", response.status_code)
    pprint(response.json())

def test_brands_list_endpoint():
    response = requests.get(f"{BASE_URL}/brands_list")
    print("Status Code:", response.status_code)
    pprint(response.json())

def test_most_similar_items_endpoint():
    params = {
        'id': 'c2191318-a71a-4d2d-9708-1c1a184941d3',
        'top_k': 10,
        'country': 'US',
        'currency': 'USD'
    }
    response = requests.get(f"{BASE_URL}/most_similar_items", params=params)
    print("Status Code:", response.status_code)
    pprint(response.json())

# def test_image_query_similarity_endpoint():
#     params = {
#         'query': 'some_image_query',
#         'img_url': 'https://static.zara.net/assets/public/075e/e927/a9954b388618/519b7a116a15/00518063802-p/00518063802-p.jpg?ts=1713514518983&w=824'
#     }
#     response = requests.get(f"{BASE_URL}/image_query_similarity", params=params)
#     print("Status Code:", response.status_code)
#     pprint(response.json())
```
